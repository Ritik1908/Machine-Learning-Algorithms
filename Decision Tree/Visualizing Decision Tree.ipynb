```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data # Four column containing sepal length, width, petal length, width
y = iris.target # Each signifies one type of flower
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

 ## Visualizing the decison tree

```python
from sklearn.tree import plot_tree
plot_tree(clf) 
```

```python
# To increase the size of tree
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80, 50
```

```python
plot_tree(clf)
```

# Visualizing For Complex Dataset:

Note: If the decision tree formed is complex take it as a warning that the decision tree is overfitted.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Dataset/Social_Network_Ads.csv")
data.head()
```

```python
# We need to change textual data to numerical
data["Gender"].replace({
        'Male': 0,
        'Female': 2
    }, inplace=True)
```

```python
data.head()
```

```python
X = data.iloc[:, 1:4].values
y = data.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(X, y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
# The above tree is condition of overfitting on data, to minimize overfitting there is a parameter called max_depth in clf.fit()
```

```python
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)
```

```python
plot_tree(clf)
```
